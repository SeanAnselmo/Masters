```python
import pandas as pd
import sqlalchemy as sq
```

```python
#engine = sq.create_engine('mysql+mysqlconnector://yourusername:yourpassword@datasciencedb.ucalgary.ca/yourusername')

with open("password.txt") as f:
    passw = f.read()

engine = sq.create_engine('mysql+mysqlconnector://katie_ovens:{}@datasciencedb.ucalgary.ca/katie_ovens'.format(passwd))
engine
```

```python
pets_df = pd.read_csv("/Users/katieovens/Downloads/Licensed_Pets.csv")
pets_df.head()
```

```python
pets_df = pets_df.rename(columns={
    'DATE': 'DATE',
    'COMMUNITY CODE': 'COMMUNITY_CODE',
    'COMMUNITY NAME': 'COMMUNITY_NAME',
    'ANIMAL': 'ANIMAL_TYPE',
    'LICENSE VOLUME': 'LICENSE_VOLUME',
    'ROW ID': 'ROW_ID',
    'POINT': 'POINT'
})
pets_df
```

```python
pets_df['DATE'] = pd.to_datetime(pets_df['DATE'], format='%m/%d/%Y %I:%M:%S %p')

pets_df.to_sql("licensed_pets", engine, index=False, if_exists='replace')
pets_table_df = pd.read_sql_table("licensed_pets", engine)
pets_table_df.head()
```

```python
select_query = "SELECT * FROM licensed_pets"
df = pd.read_sql_query(select_query, engine)
df
```

```python
select_query = "SELECT * FROM licensed_pets WHERE ANIMAL_TYPE = 'DOGS' AND LICENSE_VOLUME > 50"
df = pd.read_sql_query(select_query, engine)
df
```

```python
sql_query = """
SELECT COMMUNITY_CODE, COUNT(*) AS row_count
FROM licensed_pets
GROUP BY COMMUNITY_CODE
"""
df = pd.read_sql_query(sql_query, engine)
df
```
