## Part C: Converting from JSON to CSV (13 marks)

Write your own code, using Python built-ins (i.e. no using pandas), the csv and json modules, to convert the data in the provided file into appropriate CSVs for each of your identified elements from Part B, Question 1. You may add additional fields where necessary. Make sure there are no fields with multiple values.

Each row of each of your CSVs should contain sufficient information so that they could be used in another program independently of each other. For example, you may be able to write another program which could read your CSV file about recipes to extract information about ingredients and preparation time (maybe even what book the recipe comes from).

```python
import csv

def flatten_list(ingredients_list):
    return "; ".join(ingredients_list)

# csv lists 
categories_data = []
blogs_data = []
books_data = []
recipes_data = []

#itereating json objects to extract values 
for category in recipes['Categories']:
    category_name = category['CategoryName']
    categories_data.append([category_name])
 
    for book in category.get('Book', []):
        books_data.append([book['BookTitle'], book['ISBN'], category_name])

        for recipe in book['Recipes']:
            recipes_data.append([
                recipe['Recipe'], 
                recipe['PreparationTime'], 
                flatten_list(recipe['Ingredients']), 
                "Book: " + book['BookTitle']
            ])

            for component in recipe.get('Components', []):
                recipes_data.append([
                    component['Recipe'], 
                    component['PreparationTime'], 
                    flatten_list(component['Ingredients']), 
                    "Component of: " + recipe['Recipe']
                ])

    for blog in category.get('Blog', []):
        blogs_data.append([blog['Title'], blog['Author'], blog['Address'], category_name])

        for recipe in blog['Recipes']:
            recipes_data.append([
                recipe['Recipe'], 
                recipe['PreparationTime'], 
                flatten_list(recipe['Ingredients']), 
                "Blog: " + blog['Title']
            ])

def write_to_csv(file_path, data, header):
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(data)

categories_csv = '/Users/rs/MDataSci/data 604/hw/1/output/categories.csv'
blogs_csv = '/Users/rs/MDataSci/data 604/hw/1/output/blogs.csv'
books_csv = '/Users/rs/MDataSci/data 604/hw/1/output/books.csv'
recipes_csv = '/Users/rs/MDataSci/data 604/hw/1/output/recipes.csv'

# cols for each csv
categories_cols = ['CategoryName']
blogs_cols = ['Title', 'Author', 'Address', 'CategoryName']
books_cols = ['BookTitle', 'ISBN', 'CategoryName']
recipes_cols = ['Recipe', 'PreparationTime', 'Ingredients', 'Source']

# Writing to CSVs
write_to_csv(categories_csv, categories_data, categories_cols)
write_to_csv(blogs_csv, blogs_data, blogs_cols)
write_to_csv(books_csv, books_data, books_cols)
write_to_csv(recipes_csv, recipes_data, recipes_cols)
```
