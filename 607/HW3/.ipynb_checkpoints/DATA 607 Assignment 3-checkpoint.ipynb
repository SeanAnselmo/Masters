## 1. Can regularizing the linear model help?

In this exercise, we attempt to predict life expectance. Please find the data("LifeExpectancyData.csv") from D2L.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.utils import resample
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import warnings

# Load the dataset
data = pd.read_csv("LifeExpectancyData.csv")
data = data.dropna()

target_column = "Lifeexpectancy "
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python
data.head()
```

1. Implement the following:
   For `i` from 1 to 10000:

   - Construct a boostrap replicate `(X_tr, y_tr)` of the dataset `(X, y)`.
   - Form the complementary testing set `(X_te, y_te)` consisting of all elements of the original dataset that don't belong the bootstrap replicate.
   - Standardize `X_tr` and `X_te` using `StandardScaler`.
   - Fit
     1. A linear regression model, and
     2. A ridge regression model with parameter $\alpha=1.0$

     to the data. Record the the MSE on `(X_te, y_te)`.

2. On the same axes, plot the distributions (histograms) of the test MSES for the linear regression model and for the ridge regression model. Compute the means and standard deviations of these distributions. Indicate the means on your plot.

3. Use an appropriate statistical test to test the null hypothesis *the means of the two MSE disributions are the same*. Summarize: Is ridge regression an improvement on linear regression for this dataset?


```python
y = data['Lifeexpectancy ']
label_encoder_country = LabelEncoder()
label_encoder_status = LabelEncoder()
data['Country'] = label_encoder_country.fit_transform(data['Country'])
data['Status'] = label_encoder_country.fit_transform(data['Status'])

# Get the column names for X and Y values
target_column = "Lifeexpectancy "
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python

mse_linear = []
mse_ridge = []
n_bootstrap=10000
warnings.filterwarnings('ignore')
for i in range(n_bootstrap):
    indices = np.random.choice(len(X), round(len(X) * 0.8), replace=True)
    X_tr, y_tr = X.iloc[indices], y.iloc[indices]

    X_te = X.drop(X.index[indices])
    y_te = y.drop(y.index[indices])

    linear_reg = LinearRegression()
    linear_reg.fit(X_tr, y_tr)
    y_pred_linear = linear_reg.predict(X_te)
    mse_linear.append(mean_squared_error(y_te, y_pred_linear))

    ridge_reg = Ridge(alpha=1.0)
    ridge_reg.fit(X_tr, y_tr)
    y_pred_ridge = ridge_reg.predict(X_te)
    mse_ridge.append(mean_squared_error(y_te, y_pred_ridge))
```

```python
# Calculate average MSEs
mse_linear = np.array(mse_linear)
mse_ridge = np.array(mse_ridge)


# Calculate means and standard deviations
mean_mse_linear = np.mean(mse_linear)
std_mse_linear = np.std(mse_linear)
mean_mse_ridge = np.mean(mse_ridge)
std_mse_ridge = np.std(mse_ridge)

# Plot the distributions
plt.figure(figsize=(12, 6))
plt.hist(mse_linear, bins=50, alpha=0.5, label=f'Linear Regression\nStd: {std_mse_linear:.2f}')
plt.hist(mse_ridge, bins=50, alpha=0.5, label=f'Ridge Regression\nStd: {std_mse_ridge:.2f}')
plt.axvline(mean_mse_linear, color='blue', linestyle='dashed', linewidth=2, label=f'Linear Mean: {mean_mse_linear:.2f}')
plt.axvline(mean_mse_ridge, color='green', linestyle='dashed', linewidth=2, label=f'Ridge Mean: {mean_mse_ridge:.2f}')
plt.xlabel('MSE')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.title('Distribution of MSE for Linear and Ridge Regression')
plt.show()

# Perform t-test to compare the means of the two MSE distributions
t_stat, p_value = ttest_ind(mse_linear, mse_ridge)
print(f'T-statistic: {t_stat:.2f}, P-value: {p_value:.2e}')
```

**Question 1**

With a p value of 0.54, we fail to reject the null hypothesis that the two means of the MSE distribution are the same. We can cannot claim that means are different, which is confirmed by the V-Line for both of the means being the same on the plot. Importantly, the histograms are not the same but share the similar ranges. 

The T-stat of 0.61 further corroborates that the difference between the means of the MSEs is insignificant. We can then conclude ridge regression is not an improvement on linear regression in this dataset.

## 2. Feature engineering for time series forecasting

The goal of this exercise is to forecast demand of Australian wine. We use a dataset containing monthly sales figures between January, 1980 and July, 1995. For our convenience, I'm hosting a copy of the data on my GiHub page:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv("https://gist.githubusercontent.com/mgreenbe/8d37a66d545e132c605d5c79061f42a3/raw/9ee80a5c9f412dc31aab2ed0d59e9430a0ae97c4/australian_wine.csv")
df
```

1. In forecasting problems, it's important to reserve the **most recent** data for testing rather than splitting the data randomly into training and testing subsets. Why?

2. Train a linear regression model, predicting **red wine sales** as a function of the **time index**, i.e., the number of months after January, 1980. Train your model on the first 150 observations. Test on the remainder. Report both training and testing MSE. Plot the dataset, overlaid with your forecast. Critique your model.

3. Add a column `Year` to the dataset for *years after 1980*. For each month of the year, add a corresponding binary indicator column (consider using `pandas.get_dummies`).  Repeat 2., but regressing red wine sales onto the thirteen new columns (1 year column + 12 month indicator columns) instead of onto the time index. Comment.

**Answer to Question 2.1**

Normally, you would use the most recent portion of the time series dataset for testing, and the oldest for testing. This is because you want to ensure you are testing on the most up to date information from the dataset.

```python
df['TimeIndex'] = np.arange(len(df))

train_size = 150
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:]

# Linear regression model with TimeIndex
X_train = train_df[['TimeIndex']]
y_train = train_df['Red']
X_test = test_df[['TimeIndex']]
y_test = test_df['Red']

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)

train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f"Training MSE: {train_mse:.2f}")
print(f"Testing MSE: {test_mse:.2f}")
```

```python
# Plot the dataset, overlaid with the forecast
plt.figure(figsize=(12, 6))
plt.plot(df['YearMonth'], df['Red'], label='Actual Red Wine Sales')
plt.plot(train_df['YearMonth'], y_train_pred, label='Training Forecast')
plt.plot(test_df['YearMonth'], y_test_pred, label='Testing Forecast')
plt.xlabel('YearMonth')
plt.ylabel('Red Wine Sales')
plt.title('Red Wine Sales Forecast')
plt.legend()
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(12))  # Set to show every 12th label
plt.gca().xaxis.set_major_formatter(plt.FixedFormatter(df['YearMonth'].iloc[::12].tolist()))

plt.show()
```

**Critiques for model on Q2.2**

The MSE of our testing data is much higher than our training (394,404 compared to training's 135,383). This could mean our model is overfitting on the training data. Our forecast accurately portrays the upward trend of the model, however fails to take into account seasonality, which limits the forecast. To improve the model, incorporating features such as year will provide a model that more accurately follows the trends.

**Step 3**

```python
# Add a Year column and create binary indicator columns for each month
df['Year'] = df['YearMonth'].apply(lambda x: int(x.split('-')[0]))
df['Month'] = df['YearMonth'].apply(lambda x: int(x.split('-')[1]))
df = pd.get_dummies(df, columns=['Year', 'Month'])

# Re-split the data with new features
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:]

X_train = train_df.drop(columns=['YearMonth', 'Red', 'TimeIndex'])
y_train = train_df['Red']
X_test = test_df.drop(columns=['YearMonth', 'Red', 'TimeIndex'])
y_test = test_df['Red']

# Linear regression model with new features
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)

train_mse_new = mean_squared_error(y_train, y_train_pred)
test_mse_new = mean_squared_error(y_test, y_test_pred)

print(f"Training MSE with new features: {train_mse_new:.2f}")
print(f"Testing MSE with new features: {test_mse_new:.2f}")
```

```python
# Plot the dataset, overlaid with the new forecast
plt.figure(figsize=(12, 6))
plt.plot(df['YearMonth'], df['Red'], label='Actual Red Wine Sales')
plt.plot(train_df['YearMonth'], y_train_pred, label='Training Forecast with new features')
plt.plot(test_df['YearMonth'], y_test_pred, label='Testing Forecast with new features')
plt.xlabel('YearMonth')
plt.ylabel('Red Wine Sales')
plt.title('Red Wine Sales Forecast with new features')
plt.legend()
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(12))  # Set to show every 12th label
plt.gca().xaxis.set_major_formatter(plt.FixedFormatter(df['YearMonth'].iloc[::12].tolist()))

plt.show()
```

**Q2.3**

The orange line of the training data shows a much better fit when Year is incorporated as a feature. The green line, which represents the testing data, shows a deviation from the sample when looking a peak Red Wine Sales. With this, it is still a much better fit than the model which does not incorporate Year. The training MSE for this data is 18,375, which is substantially lower than the previous training MSE. However, the testing MSE is greater at 740,592 when compared to the previous model.

Overall, the second model is better fit to training data, however the overfitting makes the first model, on the time index, more reliable when testing on unseen data.
