```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# Define the URL of the Excel file
url = 'https://github.com/ANPC86/D607/raw/main/snake-river-plain-aqueous-chemistry-xls-1.xls'


# Read the 'WaterQuality' worksheet into a separate DataFrame
df_major = pd.read_excel(url, sheet_name='MajorDissolvedConstituents')
df_major = df_major[["Ca_mgL", "Cl_mgL", "F_mgL", "K_mgL", "Mg_mgL", "Na_mgL", "Na_K_mgL", "SiO2_mgL", "SO4_mgL", "cation_total", "anion_total"]]
def clean_data(value):
    if isinstance(value, str) and '<' in value:
        return float(value.replace('<', ''))
    return value

df_major = df_major.applymap(clean_data)
# Remove rows with "ND" values (if any)
df_major.replace("ND", pd.NA, inplace=True)
df_major.dropna(inplace=True)
df_major.head()

print(df_major)
```

```python
# Step 4: Standardize the features
features_cation_anion = ["cation_total", "anion_total"]
X_cation_anion = df_major[features_cation_anion]
scaler_cation_anion = StandardScaler()
X_cation_anion_scaled = scaler_cation_anion.fit_transform(X_cation_anion)
```

```python
mse_cation_anion = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_cation_anion_scaled)
    labels = kmeans.labels_
    mse = mean_squared_error(X_cation_anion_scaled, kmeans.cluster_centers_[labels])
    mse_cation_anion.append(mse)

optimal_k = mse_cation_anion.index(min(mse_cation_anion)) + 1

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), mse_cation_anion, marker='o', linestyle='--')
plt.title('MSE Method For Optimal k (Cation vs. Anion Total)')
plt.xlabel('Number of clusters')
plt.ylabel('MSE')
plt.show()

# Step 6: Apply K-means clustering with the optimal k
kmeans_cation_anion = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_cation_anion.fit(X_cation_anion_scaled)
df_major['Cluster_Cation_Anion'] = kmeans_cation_anion.labels_

# Step 7: Visualize the clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_major, x='cation_total', y='anion_total', hue='Cluster_Cation_Anion', palette='viridis')
plt.title('Clustering of Major Constituents Based on Cation and Anion Totals')
plt.show()

# Display the dataframe with cluster labels
print("Optimal k for Cation vs. Anion: ", optimal_k)
df_major[['cation_total', 'anion_total', 'Cluster_Cation_Anion']]
```

```python
kmeans_major = KMeans(n_clusters=2, random_state=42)
kmeans_major.fit(X_major_scaled)
df_major['Cluster_Cation_Anion'] = kmeans_major.labels_
# Step 7: Visualize the clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_major, x='cation_total', y='anion_total', hue='Cluster_Cation_Anion', palette='viridis')
plt.title('Clustering of Major Constituents Based on Cation and Anion Totals')
plt.show()

# Display the dataframe with cluster labels
print(df_major[['cation_total', 'anion_total', 'Cluster_Cation_Anion']])
```

```python
# Step 2: Keep only the specified columns
df_major = df_major[["Ca_mgL", "Cl_mgL", "F_mgL", "K_mgL", "Mg_mgL", "Na_mgL", "Na_K_mgL", "SiO2_mgL", "SO4_mgL"]]

# Step 3: Function to clean data and remove any '<' characters, keeping the numeric part
def clean_data(value):
    if isinstance(value, str) and '<' in value:
        return float(value.replace('<', ''))
    return value

# Applying the clean_data function to the dataframe
df_major = df_major.applymap(clean_data)

# Remove any remaining non-numeric values (e.g., "ND")
df_major.replace("ND", pd.NA, inplace=True)
df_major.dropna(inplace=True)

# Verify the dataframe after cleaning
print(df_major)

# Step 4: Standardize the features
features = ["Ca_mgL", "Cl_mgL", "F_mgL", "K_mgL", "Mg_mgL", "Na_mgL", "Na_K_mgL", "SiO2_mgL", "SO4_mgL"]
X_major = df_major[features]
scaler = StandardScaler()
X_major_scaled = scaler.fit_transform(X_major)

# Step 5: Determine the optimal number of clusters using the elbow method
wcss_major = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_major_scaled)
    wcss_major.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss_major, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k (Major Constituents)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Step 6: Apply K-means clustering
kmeans_major = KMeans(n_clusters=5, random_state=42)
kmeans_major.fit(X_major_scaled)
df_major['Cluster'] = kmeans_major.labels_

# Step 7: Visualize the clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_major, x='Ca_mgL', y='Cl_mgL', hue='Cluster', palette='viridis')
plt.title('Clustering of Major Constituents Based on Ca and Cl Concentrations')
plt.show()
```

```python
# Define the URL of the Excel file
url = 'https://github.com/ANPC86/D607/raw/main/snake-river-plain-aqueous-chemistry-xls-1.xls'


# Read the 'WaterQuality' worksheet into a separate DataFrame
df_major = pd.read_excel(url, sheet_name='MajorDissolvedConstituents')
df_major = df_major[["Ca_mgL", "Cl_mgL", "F_mgL", "K_mgL", "Mg_mgL", "Na_mgL", "Na_K_mgL", "SiO2_mgL", "SO4_mgL", "cation_total", "anion_total"]]
def clean_data(value):
    if isinstance(value, str) and '<' in value:
        return float(value.replace('<', ''))
    return value

df_major = df_major.applymap(clean_data)
# Remove rows with "ND" values (if any)
df_major.replace("ND", pd.NA, inplace=True)
df_major.dropna(inplace=True)
df_major.head()

print(df_major)
# Standardizing the features except for cation_total and anion_total
features = ["Ca_mgL", "Cl_mgL", "F_mgL", "K_mgL", "Mg_mgL", "Na_mgL", "Na_K_mgL", "SiO2_mgL", "SO4_mgL"]
X = df_major[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the lowest MSE
mse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    mse_value = mean_squared_error(X_scaled, kmeans.cluster_centers_[labels])
    mse.append(mse_value)

optimal_k = mse.index(min(mse)) + 1

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), mse, marker='o', linestyle='--')
plt.title('MSE Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('MSE')
plt.show()

# Apply K-means clustering with the optimal k
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
df_major['Cluster'] = kmeans.labels_

# Visualize the clusters with Ca and Cl on the axes
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_major, x='Na_mgL', y='Cl_mgL', hue='Cluster', palette='viridis')
plt.title('Clustering of Major Constituents with Na and Cl on Axes')
plt.show()

# Display the dataframe with cluster labels
print("Optimal k: ", optimal_k)
df_major[['Ba_mgL', 'Cl_mgL', 'Cluster']]
```
