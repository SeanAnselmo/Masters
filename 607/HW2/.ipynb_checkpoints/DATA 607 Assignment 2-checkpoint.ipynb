## Exercise 1: K-Means Clustering 
## Dataset : 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA


# Load the Diabetes dataset
diabetes = load_diabetes()
X = diabetes.data
```

Analyze diabetes data and provide summary details of the data set(limit your answers to couple of visualizations and a numerical summary table). Implement K-means clustering for features(X) only by fine tuning k - number of clusters. 
Apply any appropriate graphs/numerical summaries discussed in class to decide on optimal number of clusters. 

```python
#Summary Details
df = pd.DataFrame(X, columns=diabetes.feature_names)

summary = df.describe()
summary
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Silhouette Method
silhouette_avg = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    cluster_labels = kmeans.labels_
    silhouette_avg.append(silhouette_score(X_scaled, cluster_labels))

# Plotting the Silhouette Method
plt.figure(figsize=(8, 4))
plt.plot(K, silhouette_avg, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Method For Optimal k')
plt.show()
```

```python
# Applying K-means with the optimal number of clusters (let's assume k=3 from the previous analysis)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)
clusters = kmeans.labels_

# Adding cluster labels to the DataFrame
df['Cluster'] = clusters

# Summary of clusters
cluster_summary = df.groupby('Cluster').mean()
cluster_summary
```

```python
# Visualize the clusters using the first two principal components

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('K-means Clustering with {} Clusters'.format(optimal_k))

# Adding a legend
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend1)

plt.show()
```

**Findings**

Our optimal K for the sillhouette method is 3. This means that splitting the groups into 3 clusters create the best classification for the data. Further splitting the data creates undistinct groups, with clusters bleeding into one another. The practical application of this means that health care providers can investigate and create different treatment plans for different groups, and the different needs of each cluster.

## Exercise 2: Lasso and Ridge Regression


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

df=pd.read_csv("Hitters.csv")
df = df.dropna()
df = pd.get_dummies(df,dtype=int,drop_first = True)
df.head()
```

1. Split data in to train and test with 20% test data and apply standard scaling if necessary.
2. Apply Lasso regression and ridge regression with the above data split taking Salary as target variable. 
3. Optimize the model parameter and evaluate the models using mean squarred error and R^2. 
4. Comment about your finding.s 

```python
y = df.Salary
dummies = pd.get_dummies(df[['League_N', 'Division_W', 'NewLeague_N']]).astype('int')
#print(dummies.head())
X_prep = df.drop (['Salary', 'League_N', 'Division_W', 'NewLeague_N'], axis = 1)
X = pd.concat([X_prep,  dummies[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=46,test_size = 0.2)


scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Lasso with 5-fold cross-validation
lasso_model = LassoCV(cv=5, max_iter=10000, random_state=0)

# Fit model
lasso_model.fit(X_train, y_train)

# Evaluate model
y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

# Output results
lasso_model_output = pd.DataFrame({
    'alpha': [lasso_model.alpha_],
    'coeffs': [lasso_model.coef_],
    'intercepts': [lasso_model.intercept_],
    'MSE': [mse_lasso]
})
lasso_model_output['MSE'] = pd.to_numeric(lasso_model_output['MSE'])

print(f'Lasso Regression - Mean Squared Error: {mse_lasso}')
print(f'Lasso Regression - R^2 Score: {r2_lasso}')
print(f'Lasso Regression - Best Alpha: {lasso_model.alpha_}')
```

```python
# Ridge with cross-validation to find the best alpha
ridgecv = RidgeCV(alphas=(0.1, 0.2, 1.0, 10.0, 15, 20, 25), scoring='neg_mean_squared_error', cv=5)

ridgecv.fit(X_train, y_train)

best_alpha_ridge = ridgecv.alpha_
print(f'Ridge Regression - Best Alpha: {best_alpha_ridge}')

ridge_best = Ridge(alpha=best_alpha_ridge)
ridge_best.fit(X_train, y_train)
y_pred_ridge = ridge_best.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

# Output results
ridge_model_output = pd.DataFrame({
    'alpha': [best_alpha_ridge],
    'coeffs': [ridge_best.coef_],
    'intercepts': [ridge_best.intercept_],
    'MSE': [mse_ridge]
})
ridge_model_output['MSE'] = pd.to_numeric(ridge_model_output['MSE'])

print(f'Ridge Regression - Mean Squared Error: {mse_ridge}')
print(f'Ridge Regression - R^2 Score: {r2_ridge}')
```

**Findings**

The Lasso regression boasts a higher Mean square error than the Ridge regression. With this, it is only slightly larger than the Ride regression. Lasso regression also has a lower R squared value at 0.14. This means the LASSO regression method explains 14% of the residuals. LASSO sets coefficients to 0, which did not significantly improve our model in this case.

The Ridhe regressions shows slightly better performance in both MSE and R squared. This can be the better technique because if all the predictors contribute, LASSO will feature worse results. It seems this is the case, meaning this dataset and model features predictors, that to some extent, are releveant. 

## Exercise 3: Spam Detection


```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report


url = "https://raw.githubusercontent.com/mgreenbe/spam/main/spam.csv"
df = pd.read_csv(url, encoding="latin-1")
df.rename(columns={"v1": "label", "v2": "message"}, inplace=True)
df.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace=True)
df.head()
```

1. Split the data into a training set (80% of the data) and a testing set (20% of the data). Compute the proportions of ham and spam in the training set.

2. Add a feature column **length** to the dataframe `df` containing the lengths of the messages, in characters. Classify the messages in the test set as ham/spam using SVM trained on the length data.Optimize your model based on the kernel selection,Kernel parameters(if there's any) and C. Record the associated testing error to pick optimal model. 

Repeat this exercise the following two variables(two-features model): **caps**, the number of capital letters in a message, and **misc**, the number of characters in a message that aren't letters, digits, or spaces. 

3. Reconstruct SVM classifier to the (**length**, **caps**, **misc**) data. Compare its performance to those considered in part (1 & 2).


```python
# Split the data into training set (80%) and testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42, stratify=df['label'])

# Compute proportions of ham and spam in the training set
train_proportions = y_train.value_counts(normalize=True)
test_proportions = y_test.value_counts(normalize=True)
print("Training set proportions:\n", train_proportions)
print("Testing set proportions:\n", test_proportions)
```

```python
# Adding a feature column length to the dataframe
X_train = X_train.to_frame()
X_test = X_test.to_frame()
X_train['length'] = X_train['message'].apply(len)
X_test['length'] = X_test['message'].apply(len)
```

```python
# Train SVM using the length data
svm_model = make_pipeline(StandardScaler(), SVC(kernel='rbf', C=10))
svm_model.fit(X_train[['length']], y_train)

# Predict and evaluate the model
y_pred = svm_model.predict(X_test[['length']])
print("SVM with Length Feature - Accuracy:", accuracy_score(y_test, y_pred))
print("SVM with Length Feature - Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import re

# Define functions to compute additional features
def count_caps(message):
    return sum(1 for c in message if c.isupper())

def count_misc(message):
    return len(re.findall(r'[^a-zA-Z0-9\s]', message))

# Add new features to the dataframe
X_train['caps'] = X_train['message'].apply(count_caps)
X_test['caps'] = X_test['message'].apply(count_caps)
X_train['misc'] = X_train['message'].apply(count_misc)
X_test['misc'] = X_test['message'].apply(count_misc)

# Reconstruct SVM classifier using the (length, caps, misc) data
svm_model_full = make_pipeline(StandardScaler(), SVC(kernel='rbf', C=10))
svm_model_full.fit(X_train[['length', 'caps', 'misc']], y_train)

# Predict and evaluate the model
y_pred_full = svm_model_full.predict(X_test[['length', 'caps', 'misc']])
print("SVM with Length, Caps, Misc Features - Accuracy:", accuracy_score(y_test, y_pred_full))
print("SVM with Length, Caps, Misc Features - Classification Report:\n", classification_report(y_test, y_pred_full))
```

**Findings**

Look at our proportions, ham is around 87% of target variable, and the rest is classified as spam (13%). This uneveness of target variable may create inaccuracies in our model. The hyperparameter testing did not run when I tried the notebook once more, however it revealed that the kernel of 'rbf' and C=10 gave the best results. The best model was the SVM tahat incorporated Length, Caps, and Misc features. Measuring on an accuracy score basis, the model that used multiple features had an accuracy score of 93%, where the model that only used Length had an accuracy score of 88%. 

## Exercise 4: Decision trees 
The sklearn library has a lot of useful tools for tress. We first use classification trees to analyze the Carseats data set.
In these data, Sales is a continuous variable, and so we begin by recoding it as a binary variable (by thresholding). 
We use the map() function to create a variable, called High, which takes on a value of 'Y' if the Sales variable exceeds 8, 
and takes on a value of 'N' otherwise. In Python, we need to code catergorical variable into dummy variable.
"""

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz, DecisionTreeRegressor, plot_tree
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error
from sklearn import tree
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
carseats = pd.read_csv('Carseats.csv')
carseats['High'] = carseats.Sales.map(lambda x: 'Y' if x>8 else 'N')
carseats.ShelveLoc = pd.factorize(carseats.ShelveLoc)[0]
carseats.Urban = carseats.Urban.map({'No':0, 'Yes':1})
carseats.US = carseats.US.map({'No':0, 'Yes':1})
print(carseats.describe())
print(carseats.info())
print(carseats.US)
```

1. Split the data in to train and test set with 20% test size(don't forget to drop Sales and High variables).
2. Apply decision tree model by changing the arguments criteria, max_depth and min_sample_leaf. Use any appropiate methods to find tune hyperparameters. 
3. Visualize the tree based on the optimal arguments decided on step 2. 

```python
X = carseats.drop(['Sales', 'High'], axis=1)
y = carseats.High
#train_size = 200
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7,random_state=0)
```

```python
# Define the decision tree model
tree_model = DecisionTreeClassifier(random_state=0)

# Define the hyperparameters to tune
param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [None,1,3,5,7],
    'min_samples_leaf': [1, 2, 4,5,7]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=tree_model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print(f"Best Parameters: {best_params}")
print(f"Best Model: {best_model}")
```

```python
regr_tree = DecisionTreeClassifier(**best_params)
regr_tree.fit(X_train, y_train)


plt.figure(figsize=(40,20))  # customize according to the size of your tree
plot_tree(regr_tree, feature_names = X_train.columns)
plt.show()
```
